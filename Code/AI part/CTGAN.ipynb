# Load data

```python
import pandas as pd
import chardet

with open("/content/drive/MyDrive/AI Fin Fraud Detector/正式專案/CTGAN 數據生成/result.csv", "rb") as f:
    encoding = chardet.detect(f.read())["encoding"] # Detect file encoding and store in "encoding"

original_data = pd.read_csv("/content/drive/MyDrive/AI Fin Fraud Detector/正式專案/CTGAN 數據生成/result.csv", encoding=encoding) # Read file using detected encoding
original_data.head()
```

# CTGAN model processing

```python
from ctgan import CTGAN

# Count the number of rows of original data
original_data_count = len(original_data)

# Count the number of rows that need to be generated
required_synthetic_count = 10000 - original_data_count

# Get column names
columns = list(original_data.columns)

# Create a CTGAN model
ctgan = CTGAN(epochs=500)

# Train the CTGAN model
ctgan.fit(original_data, columns)

# Generate synthetic data
synthetic_data = ctgan.sample(required_synthetic_count)

synthetic_data.head()
```

# Export data into csv file

```python
# Combine original data and synthetic data
combined_data = pd.concat([original_data, synthetic_data], axis=0)

# Reset index
combined_data = combined_data.reset_index(drop=True)
combined_data["ID"] = range(1, len(combined_data) + 1)
print(len(combined_data))
print(combined_data.head())

combined_data.to_csv("/content/drive/MyDrive/AI Fin Fraud Detector/正式專案/CTGAN 數據生成/amplified_result.csv", index=False)
```
